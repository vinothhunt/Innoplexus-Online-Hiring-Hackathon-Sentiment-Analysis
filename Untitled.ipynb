```python
train['textnew'] = train['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
test['textnew'] = test['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
train['textnew'] = train['textnew'].str.replace('[^\w\s]','')
test['textnew'] = test['textnew'].str.replace('[^\w\s]','')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['textnew'] = train['textnew'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
test['textnew'] = test['textnew'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```
